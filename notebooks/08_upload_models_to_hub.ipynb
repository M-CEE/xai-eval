# Upload trained models to Hugging Face Hub

Standalone tool -- doesn't retrain or re-evaluate anything, just scans `Models/` for completed runs and uploads whichever ones aren't on the Hub yet. "Already uploaded" is determined from `model_info.json`'s `hub_commit_sha` field, so this is safe to re-run any time (e.g. after training a few new datasets/models) -- it will only touch what's new.

Requires: `pip install huggingface_hub`, and being logged in (`huggingface-cli login`, or an `HF_TOKEN` env var).
