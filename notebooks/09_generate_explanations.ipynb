# Generate explanations (SHAP + LIME)

Generates versioned SHAP and LIME explanation artifacts for each trained model. Per your partner's reproducibility requirement: explanations are generated **once** and reused -- every metric, statistical test, or figure in the paper should be computed from these saved artifacts, not from a fresh explainer run.

**Instance selection (fixed across every run):** stratified random sample of 500 test rows (or the full test set if it has fewer than 1000 rows), class balance preserved. The same selected rows are reused for every explainer -- and reused again on any future re-run of this notebook, since the selection itself is saved and loaded back rather than resampled (see `instances/metadata.json`). The same applies to the two **case instances** picked for the local-explanation plots (one correctly-classified, one incorrectly-classified test row) -- saved in `instances/case_instances.json`, shared by both SHAP's waterfall plots and LIME's standard local-explanation plots. Use `force=True` only if you deliberately want a fresh sample/pick (this breaks comparability with any explanations already generated from the old one).

**Outputs per (dataset x model) run**, under `../Explanations/<domain>/<dataset_name>/<model_name>/`:
- `shap/feature_importance.csv` -- top 10 features ranked by mean |SHAP value|
- `shap/plots/summary.png` -- beeswarm summary plot
- `shap/plots/bar.png` -- mean |SHAP value| bar ranking
- `shap/plots/waterfall_case1.png` / `waterfall_case2.png` -- correctly/incorrectly classified instance
- `lime/plots/case1.png` / `case2.png` -- LIME's standard local-explanation plot for the same two instances

Requires: `pip install shap lime`.

```python
import sys
sys.path.insert(0, "..")
from src.explanations import run_explanations
```

```python
DATASETS = [
    ("healthcare", "pima_diabetes"),
    ("healthcare", "breast_cancer_wisconsin"),
    ("healthcare", "heart_disease_uci"),
    ("finance", "loan_default"),
    ("finance", "financial_distress"),
    ("finance", "credit_card_fraud_2023"),
]
MODEL_NAMES = ["rf", "xgb"]   # matches the model_name folders under Models/<domain>/<dataset_name>/

all_results = {}

for domain, dataset_name in DATASETS:
    for model_name in MODEL_NAMES:
        print(f"\n{'='*70}\nEXPLAINING: {dataset_name} x {model_name}\n{'='*70}")

        config = {
            "dataset_name": dataset_name,
            "domain": domain,
            "model_name": model_name,
            "explainers": ["shap", "lime"],
            "datasets_root": "../Datasets",
            "models_root": "../Models",
            "explanations_root": "../Explanations",
            # everything else (random_state, small_test_threshold, n_instances,
            # lime_num_samples, lime_num_features) falls back to the fixed
            # defaults in src/explanations.py
        }

        result = run_explanations(config)
        all_results[(dataset_name, model_name)] = result
```

Each run's artifacts land under `../Explanations/<domain>/<dataset_name>/<model_name>/`:
- `instances/` -- the shared explanation subset (`metadata.json`, `X_explain.csv`, `y_explain.csv`) and the two shared case instances (`case_instances.json`)
- `shap/` -- `shap_values.npy`, `base_values.npy`, `feature_importance.csv` (top 10), `metadata.json`, `plots/` (summary, bar, two waterfalls)
- `lime/` -- `lime_explanations.csv`, `lime_instance_summary.csv`, `metadata.json`, `plots/` (two standard local-explanation plots)
- `metadata/metadata.json` -- ties it all together with the model reference (dataset, model type, hyperparameters, Hub commit if pushed)
- `logs/explanation_log.txt` -- narrative log
